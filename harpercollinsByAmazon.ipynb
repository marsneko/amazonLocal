```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
import re
```

```python
# Load the data
df = pd.read_csv('./harpercollins_date_during_1600.csv')
```

```python
# Clean the data
df.dropna(subset=['title'], inplace=True)
df.drop(columns = ['Unnamed: 0','Unnamed: 0.1'], inplace=True)
```

```python
df['pubddate'] = pd.to_datetime(df['publishDate_'])
```

```python
dfrand = pd.read_csv('./HarperCollins_random_10000.csv',encoding='unicode escape')
dfn3year = pd.read_csv('./HarperCollins_nearest_3_years.csv',encoding='unicode escape')
```

```python
fig = plt.figure(figsize=(10,6))
x = mdates.date2num(df['pubddate'])
plt.hist(x, bins = 424, alpha=0.7, label='All')
locator = mdates.AutoDateLocator()
formatter = mdates.ConciseDateFormatter(locator)
ax = plt.gca()  # Get current axis
ax.xaxis.set_major_locator(locator)
ax.xaxis.set_major_formatter(formatter)
plt.xlabel('Publication Date')
plt.ylabel('Number of Books')
plt.title('Publication Date Distribution of All Books')
plt.legend()
plt.show()
```
